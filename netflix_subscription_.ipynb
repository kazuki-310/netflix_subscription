## Find out the top 5 countries with the highest Netfrix subscriptions in 2021(Cost Per Month - Basic ($))
## Analyze why countries are divided into high and low

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')
plt.rcParams['figure.figsize'] = 10, 10

import seaborn as sns
sns.set(style="darkgrid")

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('netflix price in different countries.csv')
df.head(10)
```

```python
df.head(10)
```

```python
print(df.info())
print('------------------------------------------------')
print(df.describe())
```

```python
fig, axs = plt.subplots(2, 2, figsize=(16, 16))

sns.histplot(data=df, x="Cost Per Month - Basic ($)", kde=True, color="skyblue", ax=axs[0, 0])
sns.histplot(data=df, x="Cost Per Month - Standard ($)", kde=True, color="olive", ax=axs[0, 1])
sns.histplot(data=df, x="Cost Per Month - Premium ($)", kde=True, color="gold", ax=axs[1, 0])

plt.show()
```

```python
X = df['Cost Per Month - Basic ($)'].value_counts()
plt.title('title')
plt.pie(X, labels=X.index, autopct='%.1f%%');
```

```python
df['Value'] = pd.cut(df['Cost Per Month - Basic ($)'],3, labels=['cheep', 'medium', 'expensive|'])
```

```python
df = df.iloc[:, [0, 1 , 2, 3, 4, 7]]
```

```python
df.head(10)
```

```python
X = df['Cost Per Month - Basic ($)'].value_counts()
plt.title('title')
plt.pie(X, labels=X.index, autopct='%.1f%%');
```

### Sort by highest to lowest amount

```python
df.head()
```

```python
df_highest = df.sort_values('Cost Per Month - Basic ($)', ascending=False).head(10)
df_low = df.sort_values('Cost Per Month - Basic ($)', ascending=True).head(10)
```

```python
df_highest.head(10)
```

```python
df_low.head(10)
```

### Many of the top countries in Europe are in the northern part of the country.
### Many of these countries have high prices and high average annual incomes.

```python
sns.barplot(
    x="Cost Per Month - Basic ($)", 
    y="Country", 
    data=df_highest, 
    hue="Country",
);
```

```python
sns.barplot(
    x="Cost Per Month - Basic ($)", 
    y="Country", 
    data=df_low, 
    hue="Country",
    ci=68
);
```
